```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

```python
df_feat=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df_feat.head()
```

```python
df_feat.info()
```

```python
df_target=pd.DataFrame(cancer['target'],columns=['Cancer'])
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit(df_feat)
```

```python
scaled_feat=sc.transform(df_feat)
```

```python
df_feat_scaled=pd.DataFrame(scaled_feat,columns=df_feat.columns)
```

```python
df_feat_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_feat, np.ravel(df_target), test_size=0.30, random_state=105)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
#k=1 here
```

```python
knn.fit(X_train,y_train)
```

# to remove feature warnings we use some filterings

```python
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

## we can chane k value by trail and error method by running k from 0-40 then we can cross check accuracy with new k value

```python
error_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('Error Rate vs K-value')
plt.xlabel('K')
plt.ylabel('Error_rate')
```

```python
print('New N value =',error_rate.index(min(error_rate)))
```

```python
#with n=21 from graph
knn=KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

# example on iris dataset

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species',palette='Set2')
#setosa is most seperable from others
```

```python
# we can see what type of distribution is there by kde plot
setosa=iris[iris['species']=='setosa']
sns.kdeplot(setosa['sepal_width'],setosa['sepal_length'],cmap='plasma',shade=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit(iris.drop('species',axis=1))
```

```python
scaled_feat=sc.transform(iris.drop('species',axis=1))
```

```python
iris_feat=pd.DataFrame(scaled_feat,columns=iris.columns[:-1])
iris_feat.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( scaled_feat, iris['species'], test_size=0.30, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
